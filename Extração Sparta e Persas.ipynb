```python
##Pacotes necessários.
#!pip install openpyxl pandas re tqdm.notebook os
from openpyxl.workbook import Workbook
from openpyxl import load_workbook
from openpyxl.utils import get_column_interval
import pandas as pd
import re
import pandas as pd
from tqdm.notebook import tqdm # progress bar
from tqdm.notebook import tqdm_notebook
import os
import pandas
```

```python
## Define a função (StackOverflow) para a leitura de uma pasta
def load_workbook_range(range_string, ws):
    col_start, col_end = re.findall("[A-Z]+", range_string)

    data_rows = []
    for row in ws[range_string]:
        data_rows.append([cell.value for cell in row])

    return pd.DataFrame(data_rows, columns=get_column_interval(col_start, col_end))  
path = 'K:/DPRE/_Comum/Extração Python/Spartas e persas' ## Esse é o diretório que vai buscar as planilhas
files = []
for (dirpath, dirnames, filenames) in os.walk(path):
    files.extend(filenames) ## Pega o nome dos arquivos na pasta descrita em "path"
qtdade_arq = len(files) ## len mede a quantidade de arquivos no no path 

for i in tqdm_notebook(range(0,qtdade_arq)): ## tqdm é somente para a barra de progresso, o código rodaria  assim = "for i in range(0,qtdade_arq):"
    f = files[i] ## pega da lista file arquivo por arquivo dado i (i definido como de 0 até a quantidade de arquivo na pasta)
    wb = load_workbook(os.path.join(path, f),read_only=True,data_only=True) ## 
    if f.startswith(("SPARTA","Sparta")):
        Resultado = wb["Resultado"] 
        Capa = wb["CAPA"] 
        Efeito_med = Resultado["L10"].value ## Efeito tarifário Médio
        Tarifa_B1 = Resultado["L11"].value ## TarifaB1
        Agente = Capa["B6"].value ## Nome da distribuidora
        Data_envio = Capa["C10"].value
        Tabela = {"Agente": [Agente],'Efeito tarifário médio': [Efeito_med], 'Tarifa B1 - Residencial  (R$)': [Tarifa_B1],"Data de Envio":[Data_envio],"Tipo":"Concessionária"}
        Tabela = pd.DataFrame(Tabela)
        print(f)
    elif f.startswith(("Persas","PERSAS")):
        Resultado = wb["Resultado"] 
        Capa = wb["CAPA"] 
        Efeito_med = Resultado["L5"].value ## Efeito tarifário Médio
        Tarifa_B1 = Resultado["L6"].value ## TarifaB1
        Agente = Capa["C9"].value ## Nome da distribuidora
        Data_envio = Capa["C15"].value
        Tabela = {"Agente": [Agente],'Efeito tarifário médio': [Efeito_med], 'Tarifa B1 - Residencial  (R$)': [Tarifa_B1],"Data de Envio":[Data_envio],"Tipo":"Permissionária"}
        Tabela = pd.DataFrame(Tabela)
        print(f)
    if i == 0:
        TabelaSalva = Tabela
    elif i>0:
        Tabela = pd.concat([TabelaSalva, Tabela], ignore_index=True)
        TabelaSalva = Tabela
Tabela.to_excel("Tabela_efeito_tarifário.xlsx") ## exporta para o arquivo tabela
```
